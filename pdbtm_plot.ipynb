```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_pdbTM_rmsd(path):
    df = pd.read_csv(path)

    # 筛选X轴小于等于5的数据点
    filtered_df = df[df['RMSD'] <= 5]

    rmsd = filtered_df['RMSD']
    max_pdb_tm = filtered_df['Max_PDB_TM']
    length = filtered_df['Length']

    plt.figure(figsize=(10, 6))

    scatter = plt.scatter(rmsd, max_pdb_tm, c=length, cmap=plt.get_cmap('coolwarm'), alpha=0.7)

    cbar = plt.colorbar(scatter, label='Length')

    plt.xlabel('RMSD')
    plt.ylabel('pdbTM')
    plt.title('PDB_TM vs. RMSD (RMSD <= 5)')

    plt.show()
```

```python
path_baseline = 'full_result_baseline.csv'
plot_pdbTM_rmsd(path_baseline)
```

```python
path_baseline = 'full_result_baseline_alntmscore.csv'
plot_pdbTM_rmsd(path_baseline)
```
